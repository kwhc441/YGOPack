```python
from selenium import webdriver
#chromeのオプション設定
options = webdriver.ChromeOptions()
options.add_argument('--headless')#バックグラウンド起動
options.add_argument('--no-sandbox')#sandboxのoff
options.add_argument('--disable-dev-shm-usage')#/dev/shmパーティションの使用禁止
```

```python
cardname=input()
#ドラゴンスターにアクセス
driver=webdriver.Chrome('chromedriver',options=options)
driver.get("https://dorasuta.jp/yugioh-jp")
#検索のpopupを出す
driver.find_element_by_id("popup_search_keyword").click()
#検索欄に入力したカード名を入れる
searchbox=driver.find_element_by_name("kw2")
searchbox.send_keys(cardname)
#検索ボタンを押す
driver.find_element_by_id("search_keyword2").click()
#検索結果からカードの情報を取り出す
cards = driver.find_elements_by_class_name("element")
cardlist=[]
for i in cards:
    cardlist.append(i.text)
driver.close()
driver.quit()
for i in cardlist:
    print(i)
```
